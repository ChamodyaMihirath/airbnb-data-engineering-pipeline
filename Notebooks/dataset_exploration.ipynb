```python
from pathlib import Path
import pandas as pd

project_root = Path.cwd().resolve()
if not (project_root / "Data" / "raw" / "amsterdam").exists():
    project_root = project_root.parent

raw_path = project_root / "Data" / "raw" / "amsterdam" / "listings.csv.gz"

raw = pd.read_csv(
    raw_path,
    compression="gzip",
    keep_default_na=False,
    dtype={"price": str}
)

print("Total rows:", len(raw))
print("Price dtype:", raw["price"].dtype)
print("Null count:", raw["price"].isna().sum())
print("Empty string count:", (raw["price"] == "").sum())
print("Sample values:")
print(raw["price"].value_counts().head(10))
```
